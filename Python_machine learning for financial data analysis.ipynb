```python
import yfinance as yf
import pandas as pd

# Définir la fonction pour récupérer les données boursières
def get_stock_data(tickers, period="1y"):
    data = yf.download(tickers, period=period)['Close']
    df = pd.DataFrame(data)
    return df

# Récupérer la liste des tickers du S&P 500
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
table = pd.read_html(url)
df_tickers = table[0]
tickers = df_tickers['Symbol'].tolist()

# Télécharger les données pour les tickers du S&P 500
df_sp500 = get_stock_data(tickers, "1y")

# Calculer la moyenne des prix de clôture pour chaque ticker
average_prices = df_sp500.mean()

# Trier les tickers en fonction des cours moyens
top_tickers = average_prices.sort_values(ascending=False).head(5)

# Afficher les cinq tickers avec les cours moyens les plus élevés
print("Les cinq tickers avec les cours moyens les plus élevés :")
print(top_tickers)
```
